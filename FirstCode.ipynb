```python
import random
nombre_secret = random.randint(1,10) #ici ça cree un nombre aleatoir entre 1 et 10 et on le stock dans la variable numbre_secret
attempts = 3 # le user à 3 tantatives pour trouver cela
print("pensez a un nombre entre 1 et 10")
print("Vous avez ", attempts, "tentatives")

while attempts > 0: #tant que le user à encore les tantative on lui redemande la question jusqu'a épuisement
 guess =int(input("devinez le nombre:"))
 if guess == nombre_secret:
  print("bravo vous avez trouve le nombre")
  break
 elif guess < nombre_secret:
  print("le nombre est plus grand")
  print("il vous reste", attempts - 1, "tentatives")
  attempts -= 1 #on diminue le nombre de tentative de 1 à chaque essaie
 else:
  print("le nombre est plus petit")
  print("il vous reste", attempts - 1, "tentatives")
  attempts -= 1 #on diminue le nombre de tentative de 1 à chaque essaie

if attempts == 0:
   print("vous avez épuisé toutes vos tentatives. Le nombre secret était:",nombre_secret)
```
